#Punto 1_ Cambiar color de fondo de imagen

image.png

```python
# Definir el rango de color blanco en RGB
lower_white = np.array([200, 200, 200])
upper_white = np.array([255, 255, 255])

# Crear una mscara para los pxeles blancos
mask = cv2.inRange(imagen, lower_white, upper_white)

# Crear una copia de la imagen original para editar
imagen_green = imagen.copy()
imagen_yellow = imagen.copy()

# Cambiar los pxeles cercanos al blanco a verde en la copia verde
imagen_green[mask != 0] = [25, 255, 150]  # Color verde RGB: (112, 222, 33) equivalente a #70de21

# Cambiar los pxeles cercanos al blanco a amarillo en la copia amarilla
im=imagen_yellow[mask != 0] = [160, 255, 250]  # Color amarillo RGB: (184, 222, 33) equivalente a #b8de21

# Mostrar las tres imgenes en la misma fila
from matplotlib import pyplot as plt
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(imagen_green, cv2.COLOR_BGR2RGB))
plt.title('Pxeles Blancos a Verde')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(imagen_yellow, cv2.COLOR_BGR2RGB))
plt.title('Pxeles Blancos a Amarillo')
plt.axis('off')

plt.show()
```

#Punto 2: Aumento de brillo y contraste

##Aumento de brillo

```python
plt.figure(figsize=(20, 10))
# Original image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')
plt.axis('off')

# Image with increased brightness
Aumento_brillo = 50
imagen_bright = cv2.add(imagen, np.ones(imagen_yellow.shape, dtype=np.uint8)*Aumento_brillo)
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(imagen_bright, cv2.COLOR_BGR2RGB))
plt.title('Imagen con Brillo Aumentado')
plt.axis('off')

plt.show()

```

## Aumento de contraste

```python
plt.figure(figsize=(20, 10))
# Original image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')
plt.axis('off')

# Image with increased brightness
Aplha = 0.7 # Aumentar 1.5
Beta = -30 # Aumentar 30
imagen_bright = cv2.convertScaleAbs(imagen, alpha=Aplha, beta=Beta)
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(imagen_bright, cv2.COLOR_BGR2RGB))
plt.title('Imagen con contraste')
plt.axis('off')

plt.show()

```

#Punto 3

```python
# Sobel X
plt_x = cv2.Sobel(imagen, cv2.CV_64F, 1, 0, ksize=3)
plt_x = np.uint8(np.absolute(plt_x))

# Sobel Y
plt_y = cv2.Sobel(imagen, cv2.CV_64F, 0, 1, ksize=3)
plt_y = np.uint8(np.absolute(plt_y))

# Sobel combined
plt_xy = cv2.Sobel(imagen, cv2.CV_64F, 1, 1, ksize=3)
plt_xy = np.uint8(np.absolute(plt_xy))

# Display Sobel images
plt.figure(figsize=(20, 10))

plt.subplot(1, 4, 1)
plt.imshow(plt_x, cmap='gray')
plt.title('Sobel X')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(plt_y, cmap='gray')
plt.title('Sobel Y')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(plt_xy, cmap='gray')
plt.title('Sobel Combined')
plt.axis('off')

# Original image
plt.subplot(1, 4, 4)
plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')
plt.axis('off')

plt.show()
```
