##Importar librerias

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from PIL import Image, ImageDraw


```

##Imagen 1

```python
# Estableciendo dimiensiones
W = 500 # Ancho de la imagen
H = 500 # Altura de la imagen
# crear el un fondo cero
im = np.zeros((W,H), np.uint8) # imágenes en escala de grises con np.uint8
for i in range(H): im[i,i] = 255 # im[i,i] crea linea diagonal

cv2_imshow(im)
```

#Imagen 2

```python
W = 500
H = 500
im = np.zeros((W, H, 3), np.uint8)

# Calcula las coordenadas del centro de la imagen
center_x = W // 2
center_y = H // 2

# Dimensiones de los cuadrados
square_size = 200

# Dibuja un cuadrado azul centrado
for i in range(center_x - square_size // 2, center_x + square_size // 2):
    for j in range(center_y - square_size // 2, center_y + square_size // 2):
        im[i, j] = (0, 0, 255)  # roja

# Dibuja un cuadrado verde centrado
for i in range(center_x - square_size // 4, center_x + square_size // 4):
    for j in range(center_y - square_size // 4, center_y + square_size // 4):
        im[i, j] = (0, 255, 0)  # verde

# Dibuja un cuadrado rojo centrado
for i in range(center_x - square_size // 8, center_x + square_size // 8):
    for j in range(center_y - square_size // 8, center_y + square_size // 8):
        im[i, j] = (255, 0, 0)  # azul

# Dibuja un cuadrado negro centrado
for i in range(center_x - square_size // 16, center_x + square_size // 16):
    for j in range(center_y - square_size // 16, center_y + square_size // 16):
        im[i, j] = (0, 0, 0)  # negro

cv2_imshow(im)
```

#Imagen 3

```python
# Configuración de la imagen
W = 500
H = 500
color_bg = (0, 0, 0)  # Negro en formato RGB
color_lines = (255, 255, 255)  # Blanco en formato RGB
line_width = 1
line_spacing = 10  # Espacio entre las líneas

# Crear una nueva imagen con fondo negro del mismo tamaño que el primer ejemplo
image = Image.new("RGB", (W, H), color_bg)

# Dibujar líneas horizontales blancas delgadas en la imagen
draw = ImageDraw.Draw(image)
y = 0
while y < H:
    draw.line([(0, y), (W, y)], fill=color_lines, width=line_width)
    y += line_spacing

# Mostrar la imagen
image
```
